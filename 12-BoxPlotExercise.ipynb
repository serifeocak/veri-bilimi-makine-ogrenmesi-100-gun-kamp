```python
# Box Plotları Anlamak
# Box plots (whisker plots olarak da adlandırılır) bir veri setinin dağılımını'ını görselleştirmek için kullanılır.
# Şu konular hakkında insight sağlar:
# - Median (Q2, 50th percentile)
# - Interquartile Range (IQR) – Q1 (%25'lik yüzdelik dilim) ile Q3 (%75'lik yüzdelik dilim) arasındaki spread
# - 1.5 * IQR içindeki minimum ve maximum değerler (whiskers)
# - Outliers (whiskers dışında kalan değerler)
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,56,56,56,56,56,56,56,56,57,70])

plt.figure(figsize=(6,5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

```python
df = sns.load_dataset("titanic")
```

```python
df
```

```python
plt.figure(figsize=(15, 5))
# Box plot for Age grouped by Class
plt.subplot(1,2,1)
sns.boxplot(x="class", y="age", data=df)
plt.title("Box Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")
plt.grid(True)

# Box plot for Fare grouped by Class
plt.subplot(1,2,2)
sns.boxplot(x="class", y="fare", data=df)
plt.title("Box Plot of Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.yscale("log")  # Log scale for better visualization of high fares
plt.grid(True)

plt.tight_layout()
plt.show()
```
